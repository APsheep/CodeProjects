### Task 2: Preprocessing (10 Points)
- Do you see two columns that represent the same thing? If so, take appropriate action.
- Clean the data by removing null values, handling outliers, and addressing the redundant column.

```python
# Write your code here
def pre_process(data):
    """
    Preprocess the data
    data: Pandas DataFrame
    return: Pandas DataFrame
    """
    if 'Unnamed: 0' in data.columns:
        data = data.drop(columns=['Unnamed: 0'])
        
    if 'Day' in data.columns:
        data = data.drop(columns=['Day'])

    data['Precipitation'] = (data['Precipitation'].str.replace(r"[^\d\.]", "", regex=True).replace('', '0').astype(float)                            )
    data = data.dropna()
    data['Manhattan Bridge'] = data['Manhattan Bridge'].astype(float)
    for col in ['Brooklyn Bridge' , 'Manhattan Bridge', 'Williamsburg Bridge', 'Queensboro Bridge', 'Total']:
        data = data[(data[col] >= 0) & (data[col] <= 50000)]
    return data

    pass

# Do not modify the code below
try:
    clean_data = pre_process(data)
    display(clean_data)
    assert 'Day' not in clean_data.columns
    assert 'Unnamed: 0' not in clean_data.columns
    assert 'Date' in clean_data.columns
    assert clean_data['Precipitation'].dtype == 'float64'
    print("Your solution is correct!")
except Exception as e:
    print(f"OOPS! There is an error. {e}")
```

```python
# Write your code here
def convert_date(data):
    """
    Convert the 'Date' column to datetime format (yyyy-mm-dd)
    data: Pandas DataFrame
    return: Pandas DataFrame
    """
    data['Date'] = pd.to_datetime(data['Date'].str.strip(), format='%Y-%m-%d')
    return data
    pass

def extract_month(data):
    """
    Extract the month from the 'Date' column and add a new column 'Month'
    data: Pandas DataFrame
    return: Pandas DataFrame
    """
    data['Month'] = data['Date'].dt.month
    return data
    pass

def exctract_weekday(data):
    """
    Extract the weekday from the 'Date' column and add a new column 'Weekday'
    data: Pandas DataFrame
    return: Pandas DataFrame
    """
    data['Weekday'] = data['Date'].dt.day_name()
    return data
    pass

def drop_date(data):
    """
    Drop the 'Date' column
    data: Pandas DataFrame
    return: Pandas DataFrame
    """
    data = data.drop(columns=['Date'])
    return data
    pass

# Do not modify the code below
try:
    feature_data = convert_date(clean_data)
    feature_data = extract_month(feature_data)
    feature_data = exctract_weekday(feature_data)
    feature_data = drop_date(feature_data)
    display(feature_data)
    assert 'Month' in feature_data.columns
    assert 'Weekday' in feature_data.columns
    assert 'Date' not in feature_data.columns
    print("Your solution is correct!")
except Exception as e:    
    print(f"OOPS! There is an error. {e}")
```

```python
# Import necessary libraries
from sklearn.preprocessing import StandardScaler

def scale_data(X_train, X_test):
    """
    Scale the features using StandardScaler
    X_train: Pandas DataFrame
    X_test: Pandas DataFrame
    return: X_train_scaled, X_test_scaled
    """
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    return X_train_scaled, X_test_scaled
    pass

# Do not modify the code below
try:
    X_train_scaled, X_test_scaled = scale_data(X_train, X_test)
    assert X_train_scaled.shape == (142, 13)
    assert X_test_scaled.shape == (61, 13)
    print("Your solution might be correct!")
except Exception as e:
    print(f"OOPS! There is an error. {e}")
```

If we fit the scaler on the entire dataset (train + test), we let the model look into the test data, which would not be good for realistic and true testing since we want to test on unseen data. By fitting on the training set and transforming both sets, we ensure the same scaling logic is applied everywhere. This way we avoid having different means and std devs skewing the inputs during evaluation.
