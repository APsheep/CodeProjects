#### Task 5.3: Extracting Country Data and Creating a DataFrame (10 Points)

- Task: Extract the name, capital, population, and area for each country from the webpage and convert the data into a pandas DataFrame.

```python
# Use this cell for rough work if needed
```

```python
def extract_country_data(body):
    """
    args: body (str): a string containing the body tag data (including the body tag)
    returns: pd.DataFrame: a pandas dataframe containing the extracted data
    """
    # Complete this function
    soup = BeautifulSoup(body, "html.parser")
    countries = soup.find_all("div", class_="country")

    # Lists to store extracted data
    names = []
    capitals = []
    populations = []
    areas = []

    for country in countries:
        name = country.find("h3", class_="country-name").get_text(strip=True) if country.find("h3", class_="country-name") else None
        capital = country.find("span", class_="country-capital").get_text(strip=True) if country.find("span", class_="country-capital") else None
        population = country.find("span", class_="country-population").get_text(strip=True) if country.find("span", class_="country-population") else None
        area = country.find("span", class_="country-area").get_text(strip=True) if country.find("span", class_="country-area") else None

        names.append(name)
        capitals.append(capital)
        populations.append(population)
        areas.append(area)

    df = pd.DataFrame({
        "name": names,
        "capital": capitals,
        "population": populations,
        "area": areas
    })

    df["population"] = pd.to_numeric(df["population"].str.replace(",", ""), errors="coerce")
    df["area"] = pd.to_numeric(df["area"].str.replace(",", ""), errors="coerce")
    df["population"] = df["population"].astype(float)
    return df
    pass

# Do not change the code below

try:
    countries_df = extract_country_data(body)
    assert countries_df.shape == (250, 4), "You need to debug"
    assert countries_df.columns.tolist() == ["name", "capital", "population", "area"], "You need to debug"
    assert countries_df["area"].dtype == "float64", "Convert the area column to float"
    assert countries_df["population"].dtype == "float64", "Convert the population column to float"
    print("Test passed! You are probably correct! But verify the data with the HTML file")
    display(countries_df)
except Exception as e:
    print("OOPS! Try again\nError:", e)
```

### Task 3: Normalize (10 Points)

- Use the following conversion rates:

    - 1 INR = 0.012 USD
    - 1 CNY = 0.14 USD
    - 1 AED = 0.27 USD

```python
# Code goes here
#I put USD in here as in task 4.2 we need to remove letters anyways, so to make the process smoother I just replace the value with a float.
conversion_rates = {
    "PKR": 0.0036,
    "INR": 0.012,
    "CNY": 0.14,
    "USD": 1,
    "AED": 0.27
}

def convert_to_usd(price, currency):
    if isinstance(price, str): 
        match = re.search(r"[\d,]+", price)  
        if match:
            numeric_value = float(match.group().replace(",", ""))  
            return numeric_value * conversion_rates[currency]  
    return None  

price_columns = {
    "Launched Price (Pakistan)": "PKR",
    "Launched Price (India)": "INR",
    "Launched Price (China)": "CNY",
    "Launched Price (USA)": "USD",
    "Launched Price (Dubai)": "AED"
}

for col, currency in price_columns.items():
    df[col] = df[col].apply(lambda x: convert_to_usd(x, currency))

df.head()
#unique_values = df['Company Name'].unique()
#print(unique_values)
```
